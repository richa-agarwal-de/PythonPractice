# Aggregate

```python
rng = np.random.RandomState(0)
df4 = pd.DataFrame({
    'key':['A','B','C','A','B','C'],
    'data1': range(6),
    'data2':rng.randint(0,10,6)
},columns=['key','data1','data2'])
```

```python
df4
```

```python
#aggregate

df4.groupby('key')
```

```python
df4.groupby('key').min()
```

```python
df4.groupby('key').aggregate(['min',np.median,max])
```

```python
df4.groupby('key').aggregate('min')
```

```python
df4.groupby('key').aggregate(['min','max','mean'])
```

```python
#selective aggregate on individual columns

df4.groupby('key').aggregate({
                              'data1':'mean',
                              'data2':'sum'
                             })
```

## Filter()

```python
def func_filter(x):
  return ((x['data1'].mean() >= 2) and (x['data2'].sum() >=10))
```

```python
df4.groupby('key').filter(func_filter)
```

```python
df4.filter(lambda x: x['data2']>7) ##this doesnt work.. why????
```

```python
df4[df4['data2']>7]
```

# Transform

```python
df4['data1']**2
```

```python
df4
```

```python
df4.groupby('key').transform(lambda x: x - x.mean())
```

# Apply

```python
def norm_by_data2(x):
  x['data1'] /= x['data2'].sum()
  return x
```

```python
df4.groupby('key').apply(norm_by_data2)
```

# Specify the split key or Explicit group by 

```python
# EXPLICIT GROUP BY

L = [1,0,1,2,3,1]

df4.groupby(L).sum()
```

```python
df4.groupby(L).apply(norm_by_data2)
```

# Join/Merge

```python
dfj1 = pd.DataFrame({'employee':['Bob','Tim','Ron','Harry'],
                     'dept':['Accounting','HR','Engg','Accounting']},
                    columns=['employee','dept'])
dfj2 = pd.DataFrame({'employee':['Bob','Tim','Ron','Harry'],
                     'salary':[1000,3000,2500,2800]},
                    columns=['employee','salary'])
```

```python
dfj1
```

```python
dfj2
```

```python
dfj1.merge(dfj2)
```

```python
pd.merge(dfj1,dfj2)
```

```python
pd.merge(dfj1,dfj2,on='employee')
```

```python
pd.merge(dfj1,dfj2,how='inner',on='employee')
```

```python
dfj3 = pd.DataFrame({'name':['Bob','Harry','Hermoine'],
                     'loc':['NY','LN','CA']},
                    columns=['name','loc'])
dfj3
```

```python
pd.merge(dfj1,dfj3,left_on='employee',right_on='name',how='left')
```

```python
pd.merge(dfj1,dfj3,left_on='employee',right_on='name',how='left').drop('name',axis='columns')
```

```python
pd.merge(dfj1,dfj3,left_on='employee',right_on='name',how='right').drop('name',axis=1)
```

```python
dfj4 = dfj1.set_index('employee')
dfj4
```

```python
dfj5 = dfj3.set_index('name')
dfj5
```

```python
pd.merge(dfj4,dfj5,left_index=True,right_index=True,how='outer')
```

```python
pd.merge(dfj4,dfj3,left_index=True,right_on='name',how='outer')
```

```python
dfj6 = pd.DataFrame({'name':['Mcgonagall','Dumbledore','Snape','Harry'],
                     'dept':['Transfiguration','Headmaster','Potions','Student']}
                    ,columns=['name','dept'])
dfj6
```

```python
dfj1
```

```python
pd.merge(dfj1,dfj6,left_on='employee',right_on='name')
```

```python
pd.merge(dfj1,dfj6,left_on='employee',right_on='name',suffixes=['_l','_r'])
```

# Concat

```python
a = pd.DataFrame({'A':['A1','A2']
                  ,'B':['B1','B2']},index=[1,2],columns=['A','B'])
a
```

```python
b = pd.DataFrame({'A':['A3','A4']
                  ,'B':['B3','B4']},index=[1,2],columns=['A','B'])
b
```

```python
pd.concat([a,b])
```

```python
pd.concat([a,b],axis=1)
```

```python
pd.concat([a,b],verify_integrity=True)
```

```python
pd.concat([a,b],ignore_index=True)
```

```python
pd.concat([a,b],keys=['a','b'])
```

```python
c = pd.DataFrame({'B':['B5','B6'],
                  'C':['C5','C6']},
                 columns=['B','C'])
c
```

```python
pd.concat([a,c])
```

```python
pd.concat([a,c],join='outer')
```

```python
pd.concat([a,c],join='inner')
```

```python
pd.concat([a,c],join='outer',keys=['a','c'])
```
